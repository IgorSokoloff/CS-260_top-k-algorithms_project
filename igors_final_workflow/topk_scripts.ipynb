```python
myrepr = lambda x: repr(round(x, 4)).replace('.',',') if isinstance(x, float) else repr(x)

def read_file (shell_script_name, print_each_script):
    if print_each_script:
        f = open(shell_script_name, 'r')
        print(f.read())
        f.close()

project_path = os.getcwd() + "/"

exps = ["tournament", "quickselect"]
exps = ["quickselect"]
exps = ["tournament"]

distribution_family_ar = ["uniform", "normal"]
pivot_ar = ["random", "deterministic", "median"]

distribution_family_ar = ["normal", "uniform"]
#distribution_family_ar = ["uniform"]
#pivot_ar = ["random"]

mean = 0
sigma = 1

left = -100
right = 100

n_samples = 10_000
#n_samples = 500

d_ar = np.array([10**2, 10**3, 10**5, 10**7], dtype=int)
d_ar = np.array([10**2, 10**3, 10**4, 10**5], dtype=int)
d_ar = np.array([10**2, 10**3], dtype=int)

#d_ar = np.array([10**5], dtype=int)
num_plot_points = 10

k_ar_100 = np.linspace (max(int(0.01*100),1), 100, num_plot_points, endpoint=False, dtype=int)

time = 2
cpus_per_task = 1

if (10**5 in d_ar) or (10**6 in d_ar):
    memmory = 64
else:
    memmory = 16
slurm = 1

print_each_script = 1
print_global_script = 1

shell_script_name = "slurm_topk_quickselect_25_random"
shell_script_name = "slurm_topk_tourn_23"

single_experiment_name = "23"


print ("bash "+shell_script_name + ".sh\n")

#ef21-pp_nw-{0}_{1}x_{2}

with open(shell_script_name + ".sh", 'w') as f1:
    if "tournament" in exps:
        i = 0
        for distribution_family in distribution_family_ar:
            for i_d, d in enumerate (d_ar):
                k_ar = k_ar_100 * int(d/100)
                for i_k, k in enumerate(k_ar):
                    script_name = shell_script_name+"_"+"tournament_"+single_experiment_name +f"_{i}" + ".sh"
                    with open(script_name, 'w') as f:
                        with redirect_stdout(f):
                            print ("#!/bin/bash")
                            if distribution_family == "normal":
                                print (f"python3 synthetic_tournament_normal.py  --d {d} --k {k} --n_samples {n_samples} --mean {mean} --sigma {sigma}")
                            if distribution_family == "uniform":
                                print (f"python3 synthetic_tournament_uniform.py  --d {d} --k {k} --n_samples {n_samples} --left {left} --right {right}")
                            i += 1
                    if slurm:
                        f1.write ("sbatch --cpus-per-task={3} --mem={4}GB --time={2}:00:00 {0}{1}\n".format(project_path, script_name, time, cpus_per_task, memmory))
                    else:
                        f1.write ("bash {0}{1}\n".format(project_path, script_name))
                    read_file (script_name, print_each_script)
                    
    if "quickselect" in exps:
        i = 0
        for distribution_family in distribution_family_ar:
            for i_d, d in enumerate (d_ar):
                k_ar = k_ar_100 * int(d/100)
                for i_k, (k, pivot) in enumerate(itertools.product (k_ar, pivot_ar)):
                    script_name = shell_script_name+"_"+"quickselect_"+single_experiment_name +f"_{i}" + ".sh"
                    with open(script_name, 'w') as f:
                        with redirect_stdout(f):
                            print ("#!/bin/bash")
                            if distribution_family == "normal":
                                print (f"python3 synthetic_quickselect_normal.py  --d {d} --k {k} --n_samples {n_samples} --mean {mean} --sigma {sigma} --pivot {pivot}")
                            if distribution_family == "uniform":
                                print (f"python3 synthetic_quickselect_uniform.py  --d {d} --k {k} --n_samples {n_samples} --left {left} --right {right} --pivot {pivot}")
                            i += 1
                    if slurm:
                        f1.write ("sbatch --cpus-per-task={3} --mem={4}GB --time={2}:00:00 {0}{1}\n".format(project_path, script_name, time, cpus_per_task, memmory))
                    else:
                        f1.write ("bash {0}{1}\n".format(project_path, script_name))
                    read_file (script_name, print_each_script)

if print_global_script:    
    f = open(shell_script_name + ".sh", 'r')
    print(f.read())
    f.close()
```
